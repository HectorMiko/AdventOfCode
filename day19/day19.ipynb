```python
from pprint import pprint

with open('day19.txt') as f:
    lines = f.read()
    flows_data, inputs_data = lines.split("\n\n")

inputs = []
for i in inputs_data.split("\n"):
    d = i[1:-1].split(",")
    input = {}
    for v in d:
        key, value = v.split("=")
        input[key] = int(value)
    inputs.append(input)

pprint(inputs)
```

```python
    
flows = {}
for f in flows_data.split("\n"):
    flow_name, d = f.split('{')
    d = d[:-1]
    conditions = d.split(",")
    process = []
    for c in conditions:
        print(flow_name)
        if "<" in c:
            print("<")
            x = c.split("<")
            key, rest = x[0], "".join(x[1:])
            value, then = rest.split(":")
            condition = {
                "key" : key,
                "sign": "<",
                "value": int(value),
                "then": then
            }
        elif ">" in c:
            x = c.split(">")
            key, rest = x[0], "".join(x[1:])
            value, then = rest.split(":")
            condition = {
                "key" : key,
                "sign": ">",
                "value": int(value),
                "then": then
            }
        else:
            condition = {
                "key" : "x",
                "sign": ">",
                "value": -1,
                "then": c
            }
            
        process.append(condition)
    flows[flow_name] = process

pprint(flows)
```

```python
# inputs, flows
def compare(sign, a, b):
    if sign == ">":
        return a>b
    return a<b
def apply_flow(flow, input):
    for p in flow:
        if compare(p["sign"], input[p["key"]], p["value"]):
            return p["then"]

def input_value(input):
    return input["x"] + input["m"] + input["a"] + input["s"]
total = 0
for input in inputs:
    current_flow = "in"
    while current_flow not in ["A", "R"]:
        current_flow = apply_flow(flows[current_flow], input)
    print(current_flow)
    if current_flow=="A":
        total+= input_value(input)

print(total)
```

```python
# p2 

import copy

def get_new_flows(flow, inter):
    created_interval = []
    for p in flow:
        inf_compare = compare(p["sign"], inter[p["key"]]["inf"], p["value"])
        max_compare = compare(p["sign"], inter[p["key"]]["max"], p["value"])
        if inf_compare!=max_compare:
            #split + verify +/- 1 on splitting
            interval_1 = copy.deepcopy(inter)
            interval_2 = copy.deepcopy(inter)
            if p["sign"]=="<":
                interval_1[p["key"]]["max"] = p["value"]-1
                interval_2[p["key"]]["inf"] = p["value"]
            else:#p["sign"]==">":
                interval_1[p["key"]]["max"] = p["value"]
                interval_2[p["key"]]["inf"] = p["value"]+1
            print(p["sign"], p["value"], inter[p["key"]])
            print(interval_1[p["key"]])
            print(interval_2[p["key"]])
                
            if inf_compare:
                interval_1["flow"] = p["then"]
                created_interval.append(interval_1)
                inter = interval_2
            else:
                interval_2["flow"] = p["then"]
                created_interval.append(interval_2)
                inter = interval_1
        else :
            if inf_compare:
                inter_3 = copy.deepcopy(inter)
                inter_3["flow"] = p["then"]
                created_interval.append(inter_3)

    return created_interval

# convention : inf/max = [a, b[
current_intervals = [{
    "flow": "in",
    "x": {"inf": 1, "max": 4000},
    "m": {"inf": 1, "max": 4000},
    "a": {"inf": 1, "max": 4000},
    "s": {"inf": 1, "max": 4000},
}]
total = 0
finished_intervals = []
while len(current_intervals):
    new_current_intervals = []
    for inter in current_intervals:
        computed_intervals = get_new_flows(flows[inter["flow"]], inter)
        for computed_interval in computed_intervals:
            if computed_interval["flow"] in ["A", "R"]:
                finished_intervals.append(computed_interval)
            else:
                new_current_intervals.append(computed_interval)
    current_intervals = new_current_intervals
pprint(finished_intervals)
```

```python
total = 0
for inter in finished_intervals:
    if inter["flow"]=="A":
        total+=((inter["x"]["max"]-inter["x"]["inf"]+1)* 
                (inter["m"]["max"]-inter["m"]["inf"]+1)* 
                (inter["a"]["max"]-inter["a"]["inf"]+1)* 
                (inter["s"]["max"]-inter["s"]["inf"]+1))
print(total)
```
