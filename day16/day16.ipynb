```python
from pprint import pprint


with open('day16.txt') as f:
    lines = f.read()
    mirrors = [[char for char in line]for line in lines.split("\n")]

pprint(mirrors)
```

```python
empty_matrix = [["." for i in line] for line in mirrors]
directions = [[set() for i in line] for line in mirrors]
def change_empty_matrix(i,j,di,dj):

    if mirrors[i][j]==".":
        if di==1:
            empty_matrix[i][j] = "v"
        if di==-1:
            empty_matrix[i][j] = "^"
        if dj==1:
            empty_matrix[i][j] = ">"
        if dj==-1:
            empty_matrix[i][j] = '<'
    else: 
        empty_matrix[i][j] = mirrors[i][j]
    direction = str(di)+str(dj)
    if direction in directions[i][j]:
        return False
    else:
        directions[i][j].add(direction)
    return True
    
def follow_beam(i,j,di, dj, stack_max):
    while True:
        keep_thread = True
        if not(i<len(mirrors) and j<len(mirrors[i]) and i>=0 and j>=0):
            return
        conti = change_empty_matrix(i,j,di,dj)
        if not conti:
            return
        if mirrors[i][j] == ".":
            #continue in empty space:
            # follow_beam(i+di, j+dj, di, dj, stack_max+1)
            pass
        elif mirrors[i][j] == "/":
            if di==1:
                di = 0
                dj = -1
            elif di==-1:
                di = 0
                dj = 1
            elif dj==1:
                di = -1
                dj = 0
            elif dj==-1:
                di = 1
                dj = 0
            i = i+di
            j = j+dj
            # follow_beam(i+di, j+dj, di, dj, stack_max+1)
        elif mirrors[i][j] == "\\":
            if di==1:
                di = 0
                dj = 1
            elif di==-1:
                di = 0
                dj = -1
            elif dj==1:
                di = 1
                dj = 0
            elif dj==-1:
                di = -1
                dj = 0
            # follow_beam(i+di, j+dj, di, dj, stack_max+1)
        elif mirrors[i][j] == "|":
            if di:
                pass
                # follow_beam(i+di, j+dj, di, dj, stack_max+1)
            else:
                follow_beam(i-1, j, -1, 0, stack_max+1)
                follow_beam(i+1, j, 1, 0, stack_max+1)
                keep_thread = False
        elif mirrors[i][j] == "-":
            if dj:
                pass
                # follow_beam(i+di, j+dj, di, dj, stack_max+1)
            else:
                follow_beam(i, j+1, 0, 1, stack_max+1)
                follow_beam(i, j-1, 0, -1, stack_max+1)
                keep_thread = False
        if keep_thread:
            i = i+di
            j = j+dj
        else :
            break
        
```

```python
def count_energized(empty_matrix):
    total = 0
    for i in range(len(empty_matrix)):
        for j in range(len(empty_matrix[i])):
            if empty_matrix[i][j]!=".":
                total+=1
    return total

total = count_energized(empty_matrix)
print(total)
```

```python
max_beams = []
for i in range(len(mirrors)):
    empty_matrix = [["." for i in line] for line in mirrors]
    directions = [[set() for i in line] for line in mirrors]
    follow_beam(i,0, 0, 1, 0)
    max_beams.append(count_energized(empty_matrix))
    
    empty_matrix = [["." for i in line] for line in mirrors]
    directions = [[set() for i in line] for line in mirrors]
    follow_beam(i,len(mirrors[i])-1, 0, -1, 0)
    max_beams.append(count_energized(empty_matrix))
for j in range(len(mirrors[0])):
    try:
        empty_matrix = [["." for i in line] for line in mirrors]
        directions = [[set() for i in line] for line in mirrors]
        follow_beam(0,j, 1, 0, 0)
    except:...
    max_beams.append(count_energized(empty_matrix))
    try:
        empty_matrix = [["." for i in line] for line in mirrors]
        directions = [[set() for i in line] for line in mirrors]
        follow_beam(len(mirrors)-1, j, -1, 0, 0)
    except:...
    max_beams.append(count_energized(empty_matrix))
print(max_beams)
print(max(max_beams))
# 8357 too low
```
