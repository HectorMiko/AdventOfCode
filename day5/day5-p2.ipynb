```python
def create_ranges(src_dest_ranges) -> None:
    ranges = []
    for src_dest_range in src_dest_ranges:
        ranges.append(createNewRange(src_dest_range))
    return ranges

def createNewRange(src_dest_range):
    # print("src_dest_range", src_dest_range)
    dest, src , len = src_dest_range.split()
    return { "src": int(src), "dest": int(dest), "len": int(len), "start": int(src), "end": int(src)+int(len)-1}

def transformNumber(ranges, number_input):
    new_value = number_input
    for r in ranges:
        if (
            r["src"] <= number_input 
            and number_input<r["src"]+r["len"]
        ):
            new_value = r["dest"]+(number_input-r["src"])
    return new_value


```

```python

STEPS = ["seed-to-soil map:", "soil-to-fertilizer map:", "fertilizer-to-water map:", "water-to-light map:", "light-to-temperature map:", "temperature-to-humidity map:", "humidity-to-location map:"]
buffer_ranges = []
transformations = []
for line in data:
    if "seeds:" in line:
        seeds= [int(i) for i in line.split()[1:]]
    elif line in STEPS:
        pass
    elif len(line)==0  and len(buffer_ranges)!=0:
        print("buffer_ranges", buffer_ranges)
        t = create_ranges(buffer_ranges)
        transformations.append(t)
        del(buffer_ranges)
        buffer_ranges = []
    elif len(line):
        buffer_ranges.append(line)
t = create_ranges(buffer_ranges)
transformations.append(t)

original_seed_ranges = []
for i in range(len(seeds)):
    if i%2==1:
        continue
    original_seed_ranges.append({
        "start": int(seeds[i]),
        "end": int(seeds[i])+int(seeds[i+1]),
    })


```

```python
def split_and_transform(range_seed, transformation):
    # algo which split into n range_seed depending on transformations
    split_seeds = []
    temp_split = range_seed
    for t in transformation:
        # 5 cases : 
        # * absolutly out before
        if temp_split["end"]<t["start"]:
            split_seeds.append(temp_split)
            return split_seeds
        # * eating start
        elif temp_split["start"]<t["start"] and temp_split["end"]>=t["start"] and temp_split["end"]<=t["end"]:
            split_seeds.append({ 
                "start": temp_split["start"],
                "end": t["start"]-1
            }) 
            split_seeds.append({
                "start": t["start"],
                "end": temp_split["end"]
            })
            return split_seeds

        # * in the middle
        elif temp_split["start"]>=t["start"] and temp_split["end"]<=t["end"]:
            split_seeds.append({
                "start": temp_split["start"],
                "end": temp_split["end"]
            })
            return split_seeds
        
        # * over the total range
        elif temp_split["start"]<t["start"] and temp_split["end"]>t["end"]:
            split_seeds.append({ 
                "start": temp_split["start"],
                "end": t["start"]-1
            })
            split_seeds.append({ 
                "start": t["start"],
                "end": t["end"]
            })
            temp_split = {
                "start": t["end"]+1,
                "end": temp_split["end"]
            }

        # * eating end
        elif temp_split["start"]>=t["start"] and temp_split["start"]<=t["end"] and temp_split["end"]>t["end"]:
            split_seeds.append({ 
                "start": temp_split["start"],
                "end": t["end"]
            })
            temp_split = {
                "start": t["end"]+1,
                "end": temp_split["end"]
            }
        # * absolutly out after
        # elif temp_split["start"]>t["end"]:
        #     DO NOTHING HERE

    split_seeds.append({
        "start": temp_split["start"],
        "end": temp_split["end"]
    })
    return split_seeds

counter = 0
temp_seed_ranges = original_seed_ranges
# temp_seed_ranges = [{"start" : 50, "end": 100}]
for transformation in transformations:
    transformation = sorted(transformation, key=lambda x: x['src'])
    split_seed_list = []
    for seed in temp_seed_ranges:
        split_seed = split_and_transform(seed, transformation)
        range_size = 0
        for s in split_seed:
            if s["start"]>s["end"]:
                print("ERROR")
                print( 3/0)
            range_size+=(s["end"]-s["start"]+1)
        print(range_size, seed["end"]-seed["start"])
        if range_size!=seed["end"]-seed["start"]+1:
                print("ERROR")
                # print( 3/0)
        split_seed_list += split_seed
    transformedList = []
    for split in split_seed_list:
        transformedList.append({
            "start": transformNumber(transformation, split["start"]),
            "end": transformNumber(transformation, split["end"])
        })

    print("\n", STEPS[counter])
    temp_seed_ranges = transformedList


minimal_seed = min([s["start"] for s in temp_seed_ranges])
print(minimal_seed)
```
