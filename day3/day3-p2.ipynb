```python
with open('day3.txt') as f:
    lines = f.read()
    data = lines.split("\n")
print(data)
```

```python
#return "gear" "symbol" None
def isGear(i,j,data):
    if i<len(data) and i>=0:
        if j<len(data[i]) and j>=0:
            if data[i][j] == "*":
                return str(i)+","+str(j)

            # if data[i][j] not in "1234567890.":
            #     return {"symbol" : "symbol", "i": i, "j": j}
    return None

def findSymbolOrGear(i,j,data):
    symbols = [
        isGear(i, j+1, data),
        isGear(i, j-1, data),
        isGear(i-1, j-1, data),
        isGear(i-1, j, data),
        isGear(i-1, j+1, data),
        isGear(i+1, j-1, data),
        isGear(i+1, j, data),
        isGear(i+1, j+1, data),
    ]
    results = []
    for symbol in symbols:
        if symbol:
            results.append(symbol)
    return results
```

```python
def closeNumber(currentDigits, gears):
    if len(currentDigits) and len(gears):
        return {"number" : int(currentDigits), "gears" : gears}
    return False

parts = []
for i in range(len(data)):
    currentDigits = ""
    hasGear = False
    gears = []
    for j in range(len(data[i])):
        if data[i][j] in "1234567890":
            currentDigits+=data[i][j]
            gears += findSymbolOrGear(i,j,data)
            gears = list(set(gears))
        else : 
            part = closeNumber(currentDigits, gears)
            if (part):
               parts.append(part)
            currentDigits = ""
            gears = []
    part = closeNumber(currentDigits, gears)
    if (part):
        parts.append(part)

print(parts)
```

```python
"""reunit gears"""
def haveSameGear(part1, part2):
    commonGear = 0
    for gear in part1['gears']:
        if gear in part2["gears"]:
            commonGear+=1
    return commonGear

total = 0
for k in range(len(parts)-1):
    for l in range(k+1, len(parts)):
        commonGears = haveSameGear(parts[k], parts[l])
        if commonGears==1:
            total+=parts[k]["number"]*parts[l]["number"]
print(total)


```
