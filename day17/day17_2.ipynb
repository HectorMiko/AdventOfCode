```python
from pprint import pprint


with open('trial.txt') as f:
    lines = f.read()
    data = lines.split("\n")

pprint(data)
```

```python
INFINITY = float('inf')

class Node :
    def __init__(self, i, j, weight):
        self.shortest_paths = {}
        self.directions_path = []
        self.i = i
        self.j = j
        self.weight = int(weight)

    def is_possible_direction(self, from_direction, d, vertices):
        if self in vertices:
            return None
    
        if not len(from_direction):
            return d
        counter_last_direction = 0
        last_direction = from_direction[-1]
        for i in range(len(from_direction)):
            if from_direction[-i-1] == last_direction:
                counter_last_direction+=1
            else :
                break
        # print(from_direction, counter_last_direction)
        if last_direction == d:
            if counter_last_direction < 10:
                return (from_direction + d)[-11:]
        else :
            if counter_last_direction >= 4 :
                return (from_direction + d)[-11:]
        return None



    def set_neighbors(self, s, n, w, e):
        self.neighbors = {
            "n": n, 
            "s": s, 
            "e": e, 
            "w": w
        }

    def set_new_shortest_path(self, path, from_direction, vertex):
        if path["size"] + vertex.weight > 1500:
            return None
        if from_direction in self.shortest_paths:
            candidate_best_path = self.shortest_paths[from_direction]
            if candidate_best_path["size"] > path["size"] + vertex.weight:
                self.shortest_paths[from_direction] = {
                    "vertices": path["vertices"]+[{"node": vertex, "d": d}],
                    "size": path["size"] + vertex.weight,
                }
                return self
        else :
            self.shortest_paths[from_direction] = {
                "vertices": path["vertices"]+[{"node": vertex, "d": d}],
                "size": path["size"] + vertex.weight,
            }
            return self

    def set_shortest_path(self, path, from_direction, d, vertex):
        new_direction = self.is_possible_direction(from_direction, d, path["vertices"])
        print(from_direction, d, new_direction)
        
        if new_direction:
            return self.set_new_shortest_path(path, from_direction, vertex)
        return None


def test_nodes(nodes, i,j):
    if i<len(nodes) and i>=0:
        if j<len(nodes[i]) and j>=0:
            return True
    return False

nodes = []
for i in range(len(data)):
    nodes_line = []
    for j in range(len(data[i])):
        nodes_line.append(Node(i,j,data[i][j]))
    nodes.append(nodes_line)

for i in range(len(data)):
    for j in range(len(data[i])):
        if test_nodes(nodes, i+1, j):
            n = nodes[i+1][j]
        else:
            n = None
        if test_nodes(nodes, i-1, j):
            s = nodes[i-1][j]
        else:
            s = None
        if test_nodes(nodes, i, j+1):
            e = nodes[i][j+1]
        else:
            e = None
        if test_nodes(nodes, i, j-1):
            w = nodes[i][j-1]
        else:
            w = None
        nodes[i][j].set_neighbors(n, s, e, w)






start_node = nodes[0][0]
start_node.shortest_paths = { "" : {"vertices": [], "size": start_node.weight}}
print(INFINITY)
```
