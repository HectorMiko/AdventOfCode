```python
from pprint import pprint


with open('trial.txt') as f:
    lines = f.read()
    # data = [[c for c in line] for line in lines.split(",")]
    data = lines.split(",")

# pprint(data)
```

```python
def apply_transfo(word):
    start = 0
    for letter in word:
        start+=ord(letter)
        start*=17
        start = start%256
        
    return start

total = 0
for word in data:
    total+=apply_transfo(word)

print(total)
```

```python
print(apply_transfo("rn"))
print(apply_transfo("cm"))
print(apply_transfo("qp"))
print(apply_transfo("pc"))
```
