```python
from pprint import pprint

with open('day11.txt') as f:
    lines = f.read()
    universe = [[c for c in line] for line in lines.split("\n")]


```

```python
def is_empty_space(space):
    for i in space:
        if i != ".":
            return False
    return True

def empty_space(size):
    return ["x" for _ in range(size)]

def expand_universe(universe):
    expanded_universe = []
    for i in range(len(universe)):
        row = universe[i]
        expanded_universe.append(row)
        if is_empty_space(row):
            expanded_universe.append(empty_space(len(universe[i])))
    counter = 0
    offset = 0
    while counter<len(universe[0]):
        column = [universe[k][counter] for k in range(len(universe))]
        if is_empty_space(column):
            for i in range(len(expanded_universe)):
                expanded_universe[i] = expanded_universe[i][:counter+offset] + ['x'] + expanded_universe[i][counter+offset:] 
            offset+=1
        counter+=1
    return expanded_universe

def show(universe):
    print([str(i) for i in range(len(universe[0]))])
    for l in range(len(universe)):
        print(universe[l], str(l))
expended_universe = expand_universe(universe)
print(len(expended_universe), len(expended_universe[0]))
show(expended_universe)
```

```python
galaxies = []
for i in range(len(expended_universe)):
    for j in range(len(expended_universe[i])):
        if expended_universe[i][j]=="#":
            galaxies.append((i,j))

print(galaxies)

pairs = []
for i in range(len(galaxies)):
    for j in range(i+1, len(galaxies)):
        pairs.append((galaxies[i], galaxies[j]))
    
print(len(pairs))
```

```python
#P1

def compute_galaxies_distance(g1, g2):
    return abs(g1[0]-g2[0])+abs(g1[1]-g2[1])

total = 0
for pair in pairs:
    total+= compute_galaxies_distance(pair[0], pair[1])

print(total)
```

```python
#P1
# GAP_SIZE = 1
#P2
GAP_SIZE = 1000000

def count_row_gaps(i1, i2):
    counter = 0
    for i in range(min(i1, i2), max(i1, i2)):
        if expended_universe[i][0]=="x":
            counter+=1
    # print("count_row_gaps", counter)
    return counter
def count_column_gaps(i1, i2):
    counter = 0
    for i in range(min(i1, i2), max(i1, i2)):
        if expended_universe[0][i]=="x":
            counter+=1
    # print("count_column_gaps", counter, "range", min(i1, i2), max(i1, i2))
    return counter

def compute_galaxies_distance_p2(g1, g2):
    row_gaps = count_row_gaps(g1[0], g2[0])
    column_gaps = count_column_gaps(g1[1], g2[1])
    gaps_to_cross = row_gaps+column_gaps
    # print(g1[1], g2[1], "count_column_gaps", column_gaps)
    return abs(g1[0]-g2[0])+abs(g1[1]-g2[1]) + (gaps_to_cross) * GAP_SIZE - 2 * gaps_to_cross

total_p2 = 0
for pair in pairs:
    distance = compute_galaxies_distance_p2(pair[0], pair[1])
    # print(pair, distance)
    total_p2+= distance

total_p2
```
