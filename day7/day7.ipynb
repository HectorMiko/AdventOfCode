```python
with open('day7.txt') as f:
    lines = f.read()
    data = lines.split("\n")

# 251753912 too low
# 252489894 too low
# 252052080 too low

#P2 : 253019563 too high...
#P2 : 252898370
hands = []
for line in data : 
    h = line.split()
    hands.append({
        "cards": h[0],
        "bid": int(h[1]),
    })
```

```python
from functools import cmp_to_key


def getHandCombinationValue(cards):
    occ = {}
    counter_joker=0
    for c in cards:
        if c == "J":
            counter_joker +=1
        elif c in occ:
            occ[c] += 1
        else : 
            occ[c] = 1
    occ_values = []
    for _, value in occ.items():
        occ_values.append(value)
    occ_values = sorted(occ_values, reverse=True)

    if len(occ_values):
        max_iter = occ_values[0]+counter_joker
    else :
        max_iter = 5
    
    if max_iter == 5:
        return 6
    if max_iter == 4:
        return 5
    if max_iter == 3 and occ_values[1] == 2:
        return 4
    if max_iter == 3:
        return 3
    if max_iter == 2 and occ_values[1] == 2:
        return 2
    if max_iter == 2 :
        return 1
    return 0

def compareKicker(cards1, cards2):
    print(cards1, cards2)
    for i in range (len(cards1)):
        ki1 = "J23456789TQKA".index(cards1[i])
        ki2 = "J23456789TQKA".index(cards2[i])
        if ki1>ki2:
            return 1
        if ki1<ki2:
            return -1
    return 0

def whosTheBiggest(hand1, hand2):
    hand1_com_value = getHandCombinationValue(hand1["cards"])
    hand2_com_value = getHandCombinationValue(hand2["cards"])
    if hand1_com_value>hand2_com_value:
        return 1
    if hand1_com_value<hand2_com_value:
        return -1

    return compareKicker(hand1["cards"], hand2["cards"])

hands = sorted(hands, key=cmp_to_key(lambda item1, item2: whosTheBiggest(item1, item2)))


total = 0

for i in range(len(hands)):
    print(hands[i], i+1)
    total+= (i+1)*hands[i]["bid"]

print(total)
```
