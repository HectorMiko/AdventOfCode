```python
from pprint import pprint


with open('day14.txt') as f:
    lines = f.read()
    pattern = [[c for c in line] for line in lines.split("\n")]

# pprint(pattern)
```

```python
def transpose(pattern):
    t = [[0 for i in range(len(pattern))]for j in range(len(pattern[0]))]
    for i in range(len(pattern)):
        for j in range(len(pattern[i])):
            t[j][i] = pattern[i][j]
    return t

def roll(pattern, reverse, transposed):
    if transposed:
        to_compute_pattern = pattern
    else:
        to_compute_pattern = transpose(pattern)
    new_pattern = []
    for line in to_compute_pattern:
        sorted_splitted_lines = []
        splitted_lines = ("".join(line)).split("#")
        for splitted_line in splitted_lines:
            arr = [c for c in splitted_line]
            arr = sorted(arr)
            if reverse:
                arr = reversed(arr)
            sorted_splitted_lines.append("".join(arr))
        new_pattern.append([c for c in "#".join(sorted_splitted_lines)])
    if transposed:
        return new_pattern
    return transpose(new_pattern)

def make_them_roll(pattern, direction = "N"):
    if direction == "N":
        rolled_pattern = roll(pattern, True, False)
    if direction == "S":
        rolled_pattern = roll(pattern, False, False)
    if direction == "W":
        rolled_pattern = roll(pattern, True, True)
    if direction == "E":
        rolled_pattern = roll(pattern, False, True)
    return rolled_pattern
```

```python
rolled_pattern = make_them_roll(pattern, "S")
pprint(pattern)
pprint(rolled_pattern)
```

```python
def compute_north_load(pattern):
    total = 0
    for i in range(len(pattern)):
        for j in range(len(pattern[i])):
            if pattern[i][j]=="O":
                total+=(len(pattern)-i)
    return total

rolled_pattern = make_them_roll(pattern, "N")
value = compute_north_load(rolled_pattern)
pprint(pattern)
pprint(rolled_pattern)
print(value)
```
