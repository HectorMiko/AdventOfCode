```python
with open('day4.txt') as f:
    lines = f.read()
    data = lines.split("\n")
print(data)
```

```python
# {id : number, left: number[], right: number[]}
cards = []
for line in data:
    idTxt, numberTxt = line.split(":")
    leftNumberTxt, rightNumberTxt = numberTxt.split("|")
    cards.append({
        "number": 1,
        "id": int(idTxt.split()[-1]),
        "left": [int(n) for n in leftNumberTxt.strip().split()],
        "right": [int(n) for n in rightNumberTxt.strip().split()],
    })
print(cards)

```

```python
def computeCardP1(card):
    counter = 0
    for number in card["left"]:
        if number in card["right"]:
            counter+=1
    if counter == 0 :
        return 0
    return 2**counter/2
total = 0
for card in cards:
    total+=computeCardP1(card)

print(total)
```

```python
def computeCardP2(card):
    counter = 0
    for number in card["left"]:
        if number in card["right"]:
            counter+=1
    return counter
```

```python
cards_dict = { }
for card in cards:
    cards_dict[card["id"]] = card
print(cards_dict)
```
