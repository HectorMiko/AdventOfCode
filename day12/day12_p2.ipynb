```python
def is_possible_action(first_numbers, numbers):
    if len(first_numbers) > len(numbers):
        return False
    for i in range(len(first_numbers)):
        if i==len(first_numbers)-1:
            if first_numbers[i]<=numbers[i]:
                return False
        else :
            if first_numbers[i]!=numbers[i]:
                return False
    return True


def first_numbers_buffer(action):
    first_numbers = []
    buffer = 0
    for i in range(len(action)):
        if action[i] == "#":
            buffer+=1
        elif action[i] == "." :
            if buffer!=0:
                first_numbers.append(buffer)
                buffer = 0
        else: # c == "?"
            if buffer!=0:
                first_numbers.append(buffer)
                break
    return first_numbers

def filter_new_arr(arrs, numbers):
    #filter  arrs
    possible_arrs = []
    for i in range(len(arrs)):
        if is_possible_action(arrs[i]["numbers"], numbers):
            possible_arrs.append(arrs[i])
    dict_arrs = {}
    # group arrs by numbers/ buffer
    for i in range(len(possible_arrs)):
        id = ",".join([str(j) for j in arrs[i]["numbers"]])
        # print(id)
        if id in dict_arrs:
            dict_arrs[id]["counter"]+=arrs[i]["counter"]
        else :
            dict_arrs[id] = arrs[i]
    final_arrs = []
    for key, item in dict_arrs.items():
        final_arrs.append(item)
    return final_arrs           
    

def create_possible_arrangements(actions, numbers):
    arrs = [{"actions": actions, "numbers": [], "buffer": 0, "counter": 1}]
    for i in range(len(actions)):
        if actions[i]=="?":
            new_arr = []
            for j in range(len(arrs)):
                action_1 = [c for c in arrs[j]["actions"]]
                action_1[i] = "."
                first_numbers1 = first_numbers_buffer(action_1)
                new_arr_1 = {"actions": action_1, "numbers": first_numbers1, "counter": arrs[j]["counter"]}
                new_arr.append(new_arr_1)
                print(first_numbers1, action_1)



                action_2 = [c for c in arrs[j]["actions"]]
                action_2[i] = "#"
                first_numbers2 = first_numbers_buffer(action_2)
                print(first_numbers2, action_2)
                new_arr_2 = {"actions": action_2, "numbers": first_numbers2, "counter": arrs[j]["counter"]}
                new_arr.append(new_arr_2)
                
            filtred_new_arr = filter_new_arr(new_arr, numbers)
            arrs = filtred_new_arr

    return arrs

def compute_numbers(arr):
    numbers = []
    buffer = 0
    for c in arr:
        if c == "#":
            buffer+=1
        else :
            if buffer!=0:
                numbers.append(buffer)
                buffer = 0
    if buffer!=0:
        numbers.append(buffer)
    return numbers

def is_correct_arrangement(arr, numbers):
    computed_numbers = compute_numbers(arr)
    if len(numbers) != len(computed_numbers):
        return False
    for i in range(min(len(numbers), len(computed_numbers))):
        if numbers[i]!=computed_numbers[i]:
            return False
    return True

def compute_nb_arrangement(record):
    arrangements = create_possible_arrangements(record["record_action"], record["record_number"] )
    counter_correct_arrs = 0
    for arr in arrangements:
        print( is_correct_arrangement(arr["actions"], record["record_number"]))
        # if is_correct_arrangement(arr["actions"], record["record_number"]):
        counter_correct_arrs+=arr["counter"]
    return counter_correct_arrs




```
