```python
from pprint import pprint


with open('day18.txt') as f:
    lines = f.read()
    data = lines.split("\n")

steps = []
for step in data:
    d, n, color = step.split(" ")
    steps.append({
        "direction": d,
        "number": int(n),
        "color": color[1:-1]
    })

pprint(steps)
```

```python
canvas = [["."]]
pos_x, pos_y = 0, 0
for step in steps:
    if step["direction"] == "R":
        dx, dy = 0, 1
    elif step["direction"] == "L":
        dx, dy = 0, -1
    elif step["direction"] == "D":
        dx, dy = 1, 0
    elif step["direction"] == "U":
        dx, dy = -1, 0
    for k in range(step["number"]):
        pos_x += dx
        pos_y += dy
        if pos_x>=len(canvas):
            canvas.append(["." for j in range(len(canvas[0]))])
        elif pos_y>=len(canvas[0]):
            for j in canvas:
                j.append(".")
        elif pos_x<0:
            pos_x+=1
            canvas.insert(0, ["." for j in range(len(canvas[0]))])
        elif pos_y<0:
            pos_y+=1
            for j in canvas:
                j.insert(0, ".")
        print(pos_x)
        canvas[pos_x][pos_y] = "#"
```

```python
index = 0
for i in range(len(canvas[0])):
    if canvas[0][i]=="#":
        index = i
        break

# canvas[1][index+1] = "*"


def mark_close_node(canvas, i, j):
    coords = [[i, j]]
    counter = 0
    while len(coords):
        counter+=1
        new_coords = []
        for i, j in coords:
            if i<len(canvas) and j<len(canvas[i]):
                if canvas[i][j]==".":
                    canvas[i][j] = "#"
                    new_coords.append([i+1, j])
                    new_coords.append([i-1, j])
                    new_coords.append([i, j+1])
                    new_coords.append([i, j-1])
        coords = new_coords
mark_close_node(canvas, 1, index+1)

```

```python
for l in canvas:
    print("".join(l))
        
```
